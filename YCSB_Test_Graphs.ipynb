## Workload A: Update heavy workload
### 50% Read and 50% write

```python
import matplotlib.pyplot as plt

# Provided latency and jobs values for MySQL under "Workload A 50/50 Read/Write"
read_latency = [482.0152964, 196.5761867, 262.0135678, 164.4747723, 194.0111529]
read_jobs = [523, 5014, 25133, 49965, 100153]

write_latency = [3349.127883, 2588.040714, 3468.356295, 2797.944759, 3099.491863]
write_jobs = [477, 4986, 24867, 50035, 99847]

# Plotting the data
plt.figure(figsize=(11, 5))

# Plot for Read Latency
plt.subplot(1, 2, 1)
plt.plot(read_jobs, read_latency, 'o-', label='MySQL')
plt.title('Workload A 50/50 Read/Write - Read Latency')
plt.xlabel('Throughput (ops/sec)')
plt.ylabel('Read latency (ms)')
plt.legend()

# Plot for Write Latency
plt.subplot(1, 2, 2)
plt.plot(write_jobs, write_latency, 'o-', label='MySQL')
plt.title('Workload A 50/50 Read/Write - Write Latency')
plt.xlabel('Throughput (ops/sec)')
plt.ylabel('Write Latency (ms)')
plt.legend()

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
