# YCSB Workload Performance Analysis

In this Jupyter Notebook, we conduct a comprehensive performance analysis of three widely used database management systems: MySQL, Cassandra, and MongoDB. We utilize the Yahoo! Cloud Serving Benchmark (YCSB) to assess how each system handles a variety of workloads, ranging from heavy read operations to complex transactions. The goal is to provide insights into the latency and throughput characteristics of each database under different operational conditions, thereby aiding in the selection of the appropriate system for specific use cases.


## Workload A: Update heavy workload
### 50% Read and 50% write

```python
import matplotlib.pyplot as plt
# Workload A 50/50 Read/Write
## Read jobs and latency (us):
MySQL_rl = [482.01, 196.57, 262.01, 164.47, 194.01] #latency
MySQL_rj = [523, 5014, 25133, 49965, 100153] #jobs

MongoDB_rl = [491.08, 123.38, 75.67, 72.13, 75.29]
MongoDB_rj = [502, 4978, 24854, 50036, 100010]

Cassandra_rl = [560.75, 291.58, 284.48, 264.13, 301.31]
Cassandra_rj = [486, 4996, 25305, 49782, 99742]

## Write jobs and latency (us):
MySQL_wl = [3349.12, 2588.04, 3468.35, 2797.94, 3099.49]
MySQL_wj = [477, 4986, 24867, 50035, 99847]

MongoDB_wl = [316.39, 153.92, 96.06, 81.77, 87.76]
MongoDB_wj = [498, 5022, 25146, 49964, 99990]

Cassandra_wl = [581.33, 275.06, 248.83, 234.91, 261.75]
Cassandra_wj = [514, 5004, 24695, 50218, 100258]

# Plotting the data
plt.figure(figsize=(11, 5))

# Main title
plt.suptitle('Workload A: 50/50 Read/Write', fontsize=14, fontweight='bold')

# Plot for Read Latency
plt.subplot(1, 2, 1)
plt.plot(MySQL_rj, MySQL_rl, 'x-', label='MySQL', color='black')
plt.plot(Cassandra_rj, Cassandra_rl, 's-', label='Cassandra' ,  color='black')
plt.plot(MongoDB_rj, MongoDB_rl, 'o-', label='MongoDB',  color='black')
plt.title('Read')
plt.xlabel('Operations')
plt.ylabel('Read latency (μs)')
plt.legend()

# Plot for Write Latency
plt.subplot(1, 2, 2)
plt.plot(MySQL_wj, MySQL_wl, 'x-', label='MySQL',  color='black')
plt.plot(Cassandra_wj, Cassandra_wl, 's-', label='Cassandra',  color='black')
plt.plot(MongoDB_wj, MongoDB_wl, 'o-', label='MongoDB',  color='black')
plt.title('Update')
plt.xlabel('Operations')
plt.ylabel('Update latency (μs)')
plt.legend()

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

# YCSB Workload A: 50/50 Read/Write Performance Analysis

## Read Performance

- **MySQL**: Begins with high latency which quickly drops and stabilizes as operations increase, indicating that MySQL may require a warm-up period but then levels out, offering consistent performance at scale.
- **Cassandra**: Displays a high variance in read latency, starting high, decreasing significantly, then increasing and finally decreasing again. This fluctuation could imply that Cassandra is highly sensitive to operational load or configuration parameters.
- **MongoDB**: Shows the most consistent and lowest read latency across the number of operations. The performance remains stable as the workload increases, suggesting that MongoDB is well optimized for read operations in this scenario.

## Update Performance


- **MySQL**: Demonstrates erratic update latency with a significant peak at around 20,000 operations. However, it shows improvement with further operations, suggesting potential issues with transaction logging or buffer pool warm-up.
- **Cassandra**: Maintains a relatively low and stable update latency throughout the test. This behavior showcases Cassandra's strength in handling write operations efficiently, which is characteristic of its log-structured merge-tree storage engine.
- **MongoDB**: Exhibits the best update performance with the lowest latency, which does not significantly change as the number of operations increases. MongoDB's update performance is consistent and predictably low, reflecting its document-oriented design which allows for atomic updates on single documents.

## Summary

In this benchmark, MongoDB outperforms MySQL and Cassandra in terms of both read and update latencies, suggesting it is a strong contender for balanced workloads. However, it's important to note that benchmark results can be influenced by many factors including hardware configuration, database tuning, and the specific nature of the workload. Therefore, these results should be considered as part of a broader assessment when choosing a database for a particular application.

# Comparison of Latency vs Operations and Runtime vs Throughput

When visualizing performance metrics for database operations, the choice of what metrics to plot can significantly impact the readability and interpretability of the graphs produced. In the provided image, the top two plots depict latency in microseconds (μs) against the number of operations, while the bottom two plots show runtime in milliseconds (ms) against throughput in operations per second (ops/sec). Here’s why latency versus operations is often the more informative choice:

## Latency vs Operations

- **Clarity of Trend**: The latency metric directly measures the time taken per operation, providing a clear and immediate sense of performance per individual read/write event. This direct measurement helps in understanding the efficiency of each operation as the load increases.
- **Consistency in Scale**: The consistency in the scale of latency across different operations count allows for easier comparison between read and write operations within the same database, and also across different databases.
- **Immediate Impact Analysis**: Increases in latency are indicative of bottlenecks or inefficiencies within the database. It's easier for database administrators and engineers to correlate these spikes or trends with specific operational events or configurations.

## Runtime vs Throughput

- **Complex Interpretation**: Runtime and throughput are bulk measurements, which combine the effects of many operations. While these are important for understanding overall system performance, they can mask the behavior of individual operations.
- **Scale Variability**: Runtime can vary widely with different throughputs, leading to a less intuitive understanding of performance. This can make it challenging to deduce the performance impact of individual operations from the graphs.
- **Aggregate View**: While useful for overall system performance, the bulk view provided by runtime and throughput is less helpful for diagnosing specific performance issues. It doesn't convey the same level of detail as latency per operation, which can be crucial for optimization efforts.

In conclusion, while both pairs of plots provide valuable information, the latency vs operations plots tend to offer a more granular and immediately actionable view of database performance, which is essential for optimization and troubleshooting. They present a clearer picture of how each database system handles each operation, which is often the level of detail needed when fine-tuning systems for performance.
