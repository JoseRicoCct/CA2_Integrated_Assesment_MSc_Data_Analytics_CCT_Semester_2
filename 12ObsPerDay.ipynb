```python
from statsmodels.tsa.stattools import adfuller

# Function to perform Augmented Dickey-Fuller test
def adf_test(series):
    result = adfuller(series.dropna())  # dropna() handles any NaNs that might interfere with the test
    adf_stat = result[0]
    p_value = result[1]
    critical_values = result[4]
    return adf_stat, p_value, critical_values

# Perform ADF test on both datasets
adf_stat_full, p_value_full, critical_values_full = adf_test(dfc['vader_score'])
adf_stat_interpolated, p_value_interpolated, critical_values_interpolated = adf_test(dfc1['vader_score'])

adf_stat_full, p_value_full, critical_values_full, adf_stat_interpolated, p_value_interpolated, critical_values_interpolated
```

```python
import numpy as np


def spectral_entropy(signal, sf=1.0, method='welch', nperseg=None, normalize=False):
    """Compute the spectral entropy of a signal with the Welch method."""
    # Compute the Power Spectral Density (PSD)
    freqs, psd = welch(signal, sf, nperseg=nperseg)

    # Normalize the PSD to get a probability distribution
    psd_norm = psd / np.sum(psd)
    
    # Calculate the Shannon entropy
    entropy = -np.sum(psd_norm * np.log2(psd_norm + np.finfo(float).eps))
    
    # Optionally normalize the entropy by the entropy of a uniform distribution (to get a value between 0 and 1)
    if normalize:
        entropy /= np.log2(len(psd))
    
    return entropy

# Calculate spectral entropy for each dataset
entropy_full = spectral_entropy(dfc['vader_score'], normalize=True)
entropy_interpolated = spectral_entropy(dfc1['vader_score'], normalize=True)

entropy_full, entropy_interpolated
```

```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt


# Define train-test split point
split_point = int(len(hsa) * 0.9)  # 80% for train, 20% for test
train = hsa.iloc[:split_point]
test = hsa.iloc[split_point:]

# Fit the SARIMA model on the training data
sarima_model = SARIMAX(train['vader_score'], order=(1, 1, 1), seasonal_order=(1, 2, 1, 2))
sarima_fit = sarima_model.fit(disp=False)

# Forecast steps based on test data length
forecast_steps = 7
last_train_date = train.index[-1]

# Create forecast index starting the day after the last train date
forecast_index = pd.date_range(start=last_train_date + pd.Timedelta(days=1), periods=forecast_steps, freq='D')

# Perform a forecast
forecast_results = sarima_fit.get_forecast(steps=forecast_steps)
forecast_mean = forecast_results.predicted_mean
forecast_ci = forecast_results.conf_int()

# Plotting the results
plt.figure(figsize=(12, 6))
plt.plot(train['vader_score'], label='Training Values', color='blue')  # Training values
plt.plot(test['vader_score'], label='Test Values', color='green')  # Test values

# Connect the last training point to the first forecast point
last_train_value = train['vader_score'][-1]
first_forecast_value = forecast_mean.iloc[0]
connection_index = [last_train_date, forecast_index[0]]
connection_values = [last_train_value, first_forecast_value]
plt.plot(connection_index, connection_values, color='red', linestyle='--')  # Connection line

# Plot forecasted values
plt.plot(forecast_index, forecast_mean, label='Forecasted Values', color='red')  # Forecasted values


plt.title('Vader Score Forecast using SARIMA with Train-Test Split')
plt.xlabel('Date')
plt.ylabel('Vader Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.signal import welch

def spectral_entropy(series, sf=1.0, normalize=True):
    # Drop any NaN values from the series and use Welch's method to estimate power spectral density
    f, Pxx = welch(series.dropna(), sf)  
    # Normalize the power spectral densities so that they form a probability distribution
    psd_norm = Pxx / np.sum(Pxx)
    # Calculate the entropy of the distribution using the Shannon entropy formula
    entropy = -np.sum(psd_norm * np.log2(psd_norm + np.finfo(float).eps))
    # Normalize the entropy to a scale of 0 to 1 by dividing by the log2 of the number of points if required
    if normalize:
        entropy /= np.log2(len(psd_norm))
    return entropy

# Assuming 'btc' is your DataFrame and 'Close' is the column you want to analyze
entropy = spectral_entropy(hsa['vader_score'])
print(f'Spectral Entropy: {entropy}')
```
